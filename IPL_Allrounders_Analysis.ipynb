```python
allrounders = data.loc[(data["Player_Type"] == "Allrounder")]
#Checking Allrounders
allrounders_new = allrounders.loc[(allrounders["Capped"] == 1)]
#Taking only capped Players
Capped_Allrounders = allrounders_new[['Player_Name','Team','Nationality','Economy','Runs','Wickets','Strike_Rate']]
#Specifying the column names that you need for the selected allrounders
```

```python
Capped_Allrounders.head()
#Checking the top 5 rows
```

```python
Capped_Allrounders = Capped_Allrounders.fillna(0)
#Instead of cleaning , filling the null values with '0' , so data wont be affected and we have 
#good amount of data for analysis
```

```python
Capped_Allrounders.isna().sum()
#Checking if null values are still present after replacing with 0
```

```python
#Considering the top allrounders based on economy, Strikerate,runs and wickets
top_allrounders = Capped_Allrounders.loc[(Capped_Allrounders["Strike_Rate"] >= 132.0)]
top_allrounders_economy = top_allrounders.sort_values('Economy',ascending = False)
top_allrounders_strike_rate = top_allrounders.sort_values('Strike_Rate',ascending = False)
top_allrounders_runs = top_allrounders.sort_values('Runs',ascending = False)
top_allrounders_Wickets = top_allrounders.sort_values('Wickets',ascending = False)

```
